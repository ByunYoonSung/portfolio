```python
import re
import tokenize
import nltk
import nltk.corpus as corpus
import matplotlib as plt
import pandas as pd
import math
from nltk.corpus import brown
import numpy as np
from collections import Counter
from itertools import tee, islice
from nltk.probability import FreqDist

nltk.download('brown')
print(brown.words(),'\n')
```

```python
original_list_length = len(brown.words())
print('단어의 개수 : ',original_list_length)
```

```python
words_set = set(brown.words())
print('Unique words : \n\n',words_set,'\n')
words = brown.words()
```

```python
print('unique words의 개수 : ',len(words_set))
```

```python
#unigram의 빈도

def ngrams(n,lst) :
    tlst = lst
    while True :
        a,b = tee(tlst)
        l = tuple(islice(a,n))
        if len(l) == n : 
            yield l
            next(b)
            tlst = b
        else :
            break
            
x = Counter(ngrams(1,words))
print(x.most_common())
```

```python
y = Counter(ngrams(2,words))
print(y.most_common())
```
